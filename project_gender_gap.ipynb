### Visualizing the Gender Gap in College Majors - Dataquest Practice Project

This project focuses on visualizing the gender gap across college degrees in USA. The data covers the ratio of women bachelor level students in different degrees between the years 1970 and 2011. Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). 

First the environment is prepared and data is read in

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

Next share of female and male students is visualized for science, technology, engineering, and mathematics (STEM) degrees.

```python
cb_dark_blue = (0/255,107/255,164/255) # color for share of women
cb_orange = (255/255, 128/255, 14/255) # color for share of men
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6): # stem fields visualized in same figure in different subplots
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for value in ax.spines.values():
        value.set_visible(False)
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])

    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

Next share of female and male students is visualized for all degrees grouped in STEM, liberal arts, and other

```python
fig = plt.figure(figsize=(15, 20))

# degrees grouped into three categories:
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

print(len(stem_cats))
print(len(lib_arts_cats))
print(len(other_cats))

for sp in range(0, 6):
    ax = fig.add_subplot(6, 3, sp*3+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for value in ax.spines.values():
        value.set_visible(False)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)

    if sp == 0:
        ax.text(2007, 16, 'Men')
        ax.text(2005, 80, 'Women')
    elif sp == 5:
        ax.text(2007, 87, 'Men')
        ax.text(2005, 10, 'Women')
        ax.tick_params(labelbottom=True)

for sp in range(0, 5):
    ax = fig.add_subplot(6, 3, sp*3+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for value in ax.spines.values():
        value.set_visible(False)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)

    if sp == 0:
        ax.text(2007, 23, 'Men')
        ax.text(2005, 73, 'Women')
    elif sp == 4:
        ax.tick_params(labelbottom=True)


for sp in range(0, 6):
    ax = fig.add_subplot(6, 3, sp*3+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for value in ax.spines.values():
        value.set_visible(False)
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.set_xlim(1970, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp])
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)

    if sp == 0:
        ax.text(2007, 8, 'Men')
        ax.text(2005, 89, 'Women')
    elif sp == 5:
        ax.text(2007, 61, 'Men')
        ax.text(2005, 35, 'Women')
        ax.tick_params(labelbottom=True)

plt.savefig("gender_degrees.png")
```
