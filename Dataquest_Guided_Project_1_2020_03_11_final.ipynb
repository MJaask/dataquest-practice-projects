The google play store data has also dublicate entries for number of apps. The dublicatre 

```python
# example dublicate rows for Instagram app

for row in google_data:
    app_name = row[0]
    if app_name == 'Instagram':
        print(row)

# number of dublicate rows

duplicate_apps = []
unique_apps = []
for row in google_data:
    app_name = row[0]
    if app_name in unique_apps:
        duplicate_apps.append(app_name)
    else:
        unique_apps.append(app_name)
print('\n')
print('Number of dublicate rows', len(duplicate_apps))

# dublicate rows are removed for each app.
# The row with most reviews is assumed to be the lates data entry and will be left to the data set.

reviews_max = {}

for row in google_data:
    app_name = row[0]
    n_reviews = float(row[3])
    if app_name in reviews_max and reviews_max[app_name] < n_reviews:
        reviews_max[app_name] = n_reviews
    if app_name not in reviews_max:
        reviews_max[app_name] = n_reviews

print(reviews_max['Instagram'])
print(len(reviews_max))

android_clean = []
already_added = []

for row in google_data:
    app_name = row[0]
    n_reviews = float(row[3])
    if n_reviews == reviews_max[app_name] and app_name not in already_added:
        android_clean.append(row)
        already_added.append(row[0])

print('\n')
print('Number of rows in clean data set', len(android_clean))
print(len(already_added))
print(android_clean[:3])
```

A function to detect non-English language characters is formulated. A string/app name is considered non-English, if it has more than three non-English characters.

```python
def english_lan_check(string):
    n_non_english = []
    string = str(string)
    for character in string:
        code = ord(character)
        if code > 127:
            n_non_english.append(character)
    if len(n_non_english) > 3:
        return False
    return True

print(english_lan_check('Instagram'))
print(english_lan_check('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(english_lan_check('Docs To Go‚Ñ¢ Free Office Suite'))
print(english_lan_check('Instachat üòú'))
```

The function to identify non-English app names is used to remove all the non-English apps.

```python
english_android_apps = []
for row in android_clean:
    app_name = row[0]
    english = english_lan_check(app_name)
    if english == True:
        english_android_apps.append(row)

print('Number of apps in android data', len(android_clean))
print('Number of English language apps', len(english_android_apps))

english_apple_apps = []
for row in apple_data:
    app_name = row[1]
    english = english_lan_check(app_name)
    if english == True:
        english_apple_apps.append(row)

print('\n')
print('Number of apps in apple data',len(android_clean))
print('Number of English language apps', len(english_apple_apps))
```

## Data analysis

To minimize risks and overhead, our validation strategy for an app idea is comprised of three steps:

* Build a minimal Android version of the app, and add it to Google Play.
* If the app has a good response from users, we develop it further.
* If the app is profitable after six months, we build an iOS version of the app and add it to the App Store.

Because our end goal is to add the app on both Google Play and the App Store, we need to find app profiles that are successful on both markets.

**First most popular app genres are identified from both app stores using frequency tables:**

```python
# functions for analysis

def freq_table(dataset, index):
    frequencies = {}
    n_entries = 0
    for row in dataset:
        key = row[index]
        n_entries +=1
        if key in frequencies:
            frequencies[key] += 1
        else:
            frequencies[key] = 1
    for key in frequencies:
        frequencies[key] /= n_entries
        frequencies[key] *= 100
    return frequencies


def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
    
```

```python
# free english apple apps analysis

print(display_table(free_english_apple, 11))
print('\n')
```

Among free english apple apps, Games are the most common apps (58,2 %). After that the most common apps are Entertainment (7,9 %) and Photo & Video (5,0 %).

Games are by far the most popular category, although the results could be different, if Games would be divided into subcategories. Most apps seem to be designed for entertainment (games, photo and video, social networking, sports, music).

Based on the results, games could be a viable option, but before decision, further analysis should be done:
* what type of games are most numerous?
* which apps have most user ratings (proxy for number of users)?



```python
# free english android apps analysis

print(display_table(free_english_android, 1))
print('\n')

print(display_table(free_english_android, 9))
print('\n')
```

Among free english android apps, per category Family apps are the most common apps (18,9 %). After that the most common apps are Game (9,7 %) and Tools (8,5 %).

Based on Genre, the most common apps are Tools (8,5 %). After that the most common apps are Entertaintment apps (6,1 %) and Education apps (5,3 %).

Among anrdoid apps the seem to be more variation in number of apps per genre, and games are not clear top category as among free Apple apps. Anroid apps also seem to be less entertainment oriented compared to Apple apps.

Based on the results, no clear recommended app category or genre can be found. Further analysis should be done:

## Analysis of number of users

```python
# apple data analysis

freq_table_apple = freq_table(free_english_apple, 11)

for genre in freq_table_apple:
    total = 0
    len_genre = 0
    for row in free_english_apple:
        genre_app = row[11]
        if genre_app == genre:
            n_user_ratings = float(row[5])
            total += n_user_ratings
            len_genre += 1
    average_number_ratings = total/len_genre
    print(genre, average_number_ratings)
    
```

Based on the average number of user ratings per genre, the most used app genres are Navigation (86 000), Reference (75 000) and Social Networking (72 000). 

```python
for row in free_english_apple:
    if row[11] == 'Navigation':
        print(row[1], row[5])

print('\n')

for row in free_english_apple:
    if row[11] == 'Reference':
        print(row[1], row[5])

print('\n')

for row in free_english_apple:
    if row[11] == 'Social Networking':
        print(row[1], row[5])
```

Navigation, Reference and Social Networking genres are all skewed due to some really popular apps. Likely popular Navigation and Social Networking app development is more difficult than devoping some Reference app, so Reference app could be a genre to pursue forward.
